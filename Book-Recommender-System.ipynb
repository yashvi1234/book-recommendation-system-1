##NEW DATASET MADE BY MERGING RATINGS AND BOOKS DATASET

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

##CALCULATING THE TOTAL NUMBER OF RATING ON A SINGLE BOOK BY GROUPING IT USING IT'S BOOK-TITLE

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

##CALCULATING THE TOTAL AVERAGE RATING ON A SINGLE BOOK BY GROUPING IT USING IT'S BOOK-TITLE

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

##MERGE THE num_rating_df AND avg_rating_df on BOOK-TITLE

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

##MAKE A DATASET HAVING ONLY THOSE BOOKS ON WHICH THE TOTAL NUMBER OF RATING IS GREATER THAN 250

```python
popular_df[popular_df['num_ratings']>=250]
```

##MAKE A DATABASE IN WHICH THE TOTAL NUMBER OF RATING IS GREATER THAN 250 AND PLACE TOP 50 BOOKS IN THE DESCENDING ORDER OF THE TOTAL AVERAGE RATING ON THE BOOK

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

##MAKE A DATABASE WITH THE NAME OF THE AUTHOR, BOOK'S IMAGE AND NAME OF THE PUBLISHER FOR THESE 50 BOOKS

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommender System

##ID OF ALL THE USERS WHO HAVE RATED MORE THAN 200 TIMES, THIS IS FILTERING BASED ON THE USER

```python
z = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
user_with_high_rating = z[z].index
```

```python
requires_rating = ratings_with_name[ratings_with_name['User-ID'].isin(user_with_high_rating)]
```

##FILTERING BASED ON THE BOOKS

```python
w = requires_rating.groupby('Book-Title').count()['Book-Rating']>=50
popular_books = w[w].index
```

```python
final_ratings = requires_rating[requires_rating['Book-Title'].isin(popular_books)]
```

##INTERCHANGING THE ROWS AND COLUMN WITH EACH OTHER

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

##TO FIND THE BOOKS WITH THE MAXIMIM SIMILARITY WE ARE GOING TO FIND THE EUCLIDEAN DISTANCE 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

##MAKE A FUNCTION THAT WILL RECOMMEND THE BOOKS

```python
def recommend(book_name):
    #index fetch  from book_name
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:9]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```

```python
recommend('1984')
```

##EXPORT THE DATASETS IN PYCHARM TO ATTACH THE DATASET TO THE WEBSITE

```python
import pickle
with open('popular','wb') as pop:
    pickle.dump(popular_df,pop)

    
with open('pts','wb') as pop:
    pickle.dump(pt,pop)

    
with open('similarity_socres','wb') as pop:
    pickle.dump(similarity_scores,pop)
    
    
with open('bks','wb') as pop:
    pickle.dump(books,pop)
```
